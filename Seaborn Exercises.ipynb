# Importing Library

```python
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

# Loading Dataset

```python
df = sns.load_dataset('tips')
df.head()
```

# Distribution Plots 

# Correlation Heatmap

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

# Joint Plot

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

# Pair Plot 

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

# Dist Plot

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

# Categorical Plots

# Count Plot

```python
sns.countplot(x='sex',data=df)
```

```python
sns.countplot(x='smoker',data=df)
```

```python
sns.countplot(x='day',data=df)
```

### Horizontal Count Plot

```python
sns.countplot(y='sex',data=df)
```

# Bar Plot

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
sns.barplot(x='smoker',y='total_bill',data=df)
```

# Box Plot

```python
sns.boxplot(x='smoker',y='total_bill',data=df)
```

```python
sns.boxplot(x='sex',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(x='day',y='total_bill',hue='smoker',data=df)
```

# Violin Plot

```python
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
